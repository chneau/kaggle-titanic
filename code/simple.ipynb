```python
import warnings
warnings.simplefilter("ignore")
import pandas as pd
import re
from keras import Model
from keras.layers import Dense
from keras.layers import Dropout
from keras_pandas.Automater import Automater
```

```python
df = pd.read_csv(r"../input/train.csv")
def add_title(cell): return re.search(' ([A-Za-z]+)\\.', cell).group(1)
df["Title"] = df.apply(lambda row: add_title(row["Name"]), axis=1)
df[:5]
```

```python
auto = Automater(categorical_vars=['Pclass', 'Sex', 'Survived'],
                 numerical_vars=['Age', 'Fare', 'SibSp', 'Parch'],
                 text_vars=['Title', 'Cabin'],
                 response_var='Survived')
X, y = auto.fit_transform(df) # Transform the data set, using keras_pandas

x = auto.input_nub # Start model with provided input nub

x = Dense(16)(x) 
# x = Dropout(0.2)(x) 
# x = Dense(16, activation='relu')(x)
# x = Dense(16, activation='softmax')(x)
# x = Dense(16, activation='elu')(x)
# x = Dense(16, activation='selu')(x)
# x = Dense(16, activation='softplus')(x)
# x = Dense(16, activation='tanh')(x)
# x = Dense(16, activation='sigmoid')(x)
# x = Dense(16, activation='hard_sigmoid')(x)
# x = Dense(16, activation='linear')(x)
# x = Dense(16)(x) # Fill in your own hidden layers

x = auto.output_nub(x) # End model with provided output nub

model = Model(inputs=auto.input_layers, outputs=x)
model.compile(optimizer='Adam', loss=auto.loss, metrics=['accuracy'])

model.fit(X, y, epochs=25, validation_split=.2, verbose=2) # Train model
```
